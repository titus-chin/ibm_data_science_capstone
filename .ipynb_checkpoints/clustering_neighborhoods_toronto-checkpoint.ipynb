## Part 1: Scraping and Cleaning Data

```python
# import necessary library
import numpy as np
import pandas as pd

# change the option to display all columns and rows
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# get the wikipedia link and set it as variable url
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

# use pandas.read_html function to scrape tables from url
df=pd.read_html(url)

# shows the number of tables
print(len(df))
```

```python
# the table we are interested in is the first table
df=df[0]
df
```

```python
# replace 'Not assigned' with numpy.nan
df.replace('Not assigned',np.nan,inplace=True)

# drop the rows if the Borough = numpy.nan 
df.dropna(subset=['Borough'], inplace=True)

# reset index
df.reset_index(drop=True, inplace=True)

# show the whole dataframe
df
```

```python
# display the number of rows of dataframe
df.shape[0]
```

## Part 2: Adding Latitudes and Longitudes to Dataframe

```python
# import necessary library
import geocoder
```

```python
# using arcgis API to get the latitudes and longitudes of each postal code
latitude=[]
longitude=[]
for code in df['Postal Code']:
    g = geocoder.arcgis('{}, Toronto, Ontario'.format(code))
    print(code, g.latlng)
    while (g.latlng is None):
        g = geocoder.arcgis('{}, Toronto, Ontario'.format(code))
        print(code, g.latlng)
    latlng = g.latlng
    latitude.append(latlng[0])
    longitude.append(latlng[1])
```

```python
# add latitude and longitude to dataframe
df['Latitude']=latitude
df['Longitude']=longitude
df
```

## Part 3: Segmenting and Clustering Neighborhoods in Toronto

In this part, I am going to segment and cluster all the neighborhoods in Toronto into 5 clusters according to the common venues nearby the neighborhoods recommended by the Foursquare API.

```python
# import necessary libraries
import json
import requests
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
import folium
```

```python
# get the latitude and longitude of toronto city
g1 = geocoder.arcgis('Toronto')

# initialize a map of toronto city
map_toronto = folium.Map(location=g1.latlng,zoom_start=10)

# add all neighborhoods as markers into the map
for lat, lng, borough, code, neighbourhood in zip(df['Latitude'],df['Longitude'], df['Borough'], df['Postal Code'], df['Neighbourhood']):
    label = '{}, {}, {}'.format(code, neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  

# show the map of toronto city    
map_toronto
```

```python
# set the foursquare API credentials
# in order to run this notebook, you need to enter your own foursquare API credentials
CLIENT_ID = 'Your ID'
CLIENT_SECRET = 'Your Secret'
VERSION = '20201102'
LIMIT = 100
```

```python
# define a function to explore top 100 venues for each neighborhood
def getNearbyVenues(codes, latitudes, longitudes, radius=500):
    venues_list=[]
    for code, lat, lng in zip(codes, latitudes, longitudes):
        print(code) 
        
        # create the foursquare API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)  
        
        # make the GET request to get the top 100 venues
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each venue
        venues_list.append([(
            code,
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])
    
    # transform venues_list into a dataframe
    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Postal Code', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    return(nearby_venues)
```

```python
# use getNearbyVenues function to obtain top 100 venues for each neighborhood
top_venues = getNearbyVenues(codes=df['Postal Code'],
                             latitudes=df['Latitude'],
                             longitudes=df['Longitude']
                            )
```

```python
# apply one hot encoding method to the 'Venue Category' column
toronto_onehot = pd.get_dummies(top_venues[['Venue Category']], prefix="", prefix_sep="")

# add postal code column to toronto_onehot dataframe
toronto_onehot['Postal Code'] = df['Postal Code'] 

# move postal code column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

# print the number of rows and columns in toronto_onehot dataframe
print(toronto_onehot.shape)

# show the first 5 rows of dataframe
toronto_onehot.head()
```

```python
# group the data by Postal Code
toronto_grouped = toronto_onehot.groupby('Postal Code').mean().reset_index()

# print the number of rows and columns in toronto_grouped dataframe
print(toronto_grouped.shape)

# show the dataframe
toronto_grouped
```

```python
# define a function to show the most common venues
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
# create columns for top 10 venues
num_top_venues = 10
indicators = ['st', 'nd', 'rd']
columns = ['Postal Code']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# add data into the dataframe
postalcode_venues_sorted = pd.DataFrame(columns=columns)
postalcode_venues_sorted['Postal Code'] = toronto_grouped['Postal Code']
for ind in np.arange(toronto_grouped.shape[0]):
    postalcode_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :],num_top_venues)

# show the first 5 rows of dataframe    
postalcode_venues_sorted.head()
```

```python
# set number of clusters to 5
kclusters = 5

# drop the 'Postal Code' column
toronto_grouped_clustering = toronto_grouped.drop('Postal Code', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for first 10 rows in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels into the dataframe
postalcode_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

# merge all the data into a dataframe called toronto_df
toronto_df = df
toronto_df = toronto_df.join(postalcode_venues_sorted.set_index('Postal Code'), on='Postal Code')

# show first 5 rows of dataframe
toronto_df.head()
```

```python
# initialize a map for toronto city again
toronto_map_clusters = folium.Map(location=g1.latlng, zoom_start=10)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, code, cluster, borough,neighborhood in zip(toronto_df['Latitude'],
                                                         toronto_df['Longitude'],
                                                         toronto_df['Postal Code'],
                                                         toronto_df['Cluster Labels'],
                                                         toronto_df['Borough'],
                                                         toronto_df['Neighbourhood']):
    label = '{}, {}, {}. Cluster: {}'.format(code, neighbourhood, borough, cluster)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(toronto_map_clusters)

# show the toronto_map_clusters    
toronto_map_clusters
```
