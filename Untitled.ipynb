```python
import requests
from pandas.io.json import json_normalize
```

```python
url = 'https://services.arcgis.com/S9th0jAJ7bqgIRjw/arcgis/rest/services/Neighbourhood_MCI/FeatureServer/0/query?where=1%3D1&outFields=Neighbourhood,Hood_ID,Population,Assault_AVG,AutoTheft_AVG,Homicide_AVG,TheftOver_AVG,BreakandEnter_AVG,Robbery_AVG&outSR=4326&f=json'
results = requests.get(url).json()
```

```python
crime_data = results['features']
dataframe = json_normalize(crime_data)
dataframe
```

```python
import pandas as pd
df_temp = pd.DataFrame(columns=['ID','Neighborhood'])
```

```python
df_temp['ID']=dataframe['attributes.Hood_ID']
df_temp['Neighborhood']=dataframe['attributes.Neighbourhood']
```

```python
df_temp.sort_values(by='ID',inplace=True)
```

```python
df_temp.reset_index(drop=True,inplace=True)
```

```python
df_temp
```

```python
df_temp['Neighborhood'].replace(' ', '-',regex=True, inplace=True)
df_temp.head(20)
```

```python
df_temp['Neighborhood']=df_temp['Neighborhood'].str.lower()
df_temp
```

```python
import requests 
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.keys import Keys
import time
import numpy as np

driver = webdriver.Firefox()
neighborhoods=[]
url='https://www.realosophy.com/etobicoke-former-toronto/neighbourhood-map'
driver.get(url)
time.sleep(5)
html= driver.page_source
soup = BeautifulSoup(html,'lxml')
all_divs = soup.find('div',{'class':'row mt-4'})
neighborhoods_data = all_divs.find_all('a')
for neighborhood in neighborhoods_data : 
    neighborhoods.append(neighborhood.text)
driver.quit()
neighborhoods
```

```python
websites=[]
for website in neighborhoods_data:
    websites.append(website['href'])
websites
```

```python
import requests 
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.keys import Keys
import time
import numpy as np

driver = webdriver.Firefox()

df_temp['House Price']=0

for i in range(20):
    url='https://www.realosophy.com/{}-former-toronto/neighbourhood-map?neighbourhood={}-toronto'.format('etobicoke', df_temp['Neighborhood'][i])
    driver.get(url)
    time.sleep(10)
    html= driver.page_source
    
    soup = BeautifulSoup(html,'lxml')
    all_divs = soup.find('div',{'class':'neighbourhood-map__avg-sale-price ng-binding ng-scope'})        
    
    if all_divs == None:
        df_temp['House Price'][i] = np.nan
        print('{}: {}'.format(df_temp['Neighborhood'][i],df_temp['House Price'][i]))
    else:
        HousePrice = all_divs.text.strip()
        HousePrice = HousePrice.replace('$','')
        if 'M' in HousePrice:
            HousePrice=HousePrice.replace('M','')
            df_temp['House Price'][i]=int(float(HousePrice)*1000000)
            print('{}: {}'.format(df_temp['Neighborhood'][i],df_temp['House Price'][i]))
        else:
            HousePrice=HousePrice.replace('K','')
            df_temp['House Price'][i]=int(HousePrice)*1000
            print('{}: {}'.format(df_temp['Neighborhood'][i],df_temp['House Price'][i]))

driver.quit()
```
